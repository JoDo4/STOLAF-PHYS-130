### Part 1: Setup the scene and display

The chunk of code below is simply creating the canvas (window) that we will be interacting with (i.e., creating our spheres and displaying animations). We will specify the name, the title, the width, the height, and the background of the canvas. For more information, look at the documentation of the canvas object [here](https://www.glowscript.org/docs/VPythonDocs/canvas.html).

```python
from vpython import *
# REMEMBER TO ACTIVATE THE VIRTUAL ENVIRONMENT phyenv
current = 0
current = canvas.get_selected()
if current != 0:
    current.delete()
# --------------------
# Scene setup 
# --------------------
intro_scene = canvas() # set the variable intro_scene to "hold" our canvas
intro_scene.title = "Starter code" # set the title
intro_scene.width = 800 # specify the width
intro_scene.height = 600 # specify the height
intro_scene.background = color.black # set the background to the desired color

### OUTPUT WILL ALWAYS BE SHOWN BELOW THIS CELL
```

### Part 2: Declare our constants

This is where we will define variables so we can use them later in the program. For example, when calculating the gravitational force on an object, there is the gravitational constant (G), so we need to define its value in our code. We will also set up the "zoom" of the canvas, so how much we on screen, and set up the scale factor to go from physical units to screen units.

```python

# --------------------
# constants
# -------------------
G = 6.7e-11
mEarth = 6e24
rEarth = 6.4e6
mcraft = 15e3
planetloc = vector(0,0,0)
craftloc = vector(-10*rEarth, 0, 0)

# define all constants relating to time

t = 0
dt = 60
frames = 100
time_limit = 10*365*24*60*60 # Provide a limit for the while loop to stop

# --------------------
# display setup
# --------------------
displaymax = 10
intro_scene.range = displaymax

physicalmax = mag(craftloc)
scalefactor = displaymax/physicalmax
```

### Part 3: Create and initialize our objects

Now that our canvas is set up, we need to put the Earth and the craft on the screen. Below, we create spheres for both the Earth and the craft, setting the position to what we had in the previous part, multiplied by scalefactor, so everything is scaled to the right size. We also set the initial velocity and momentum of the craft.

```python


# --------------------
# create display objects
# --------------------

Earth = sphere(
    pos = planetloc, 
    radius = rEarth*scalefactor, 
    texture = "https://i.imgur.com/do31Gdf.jpg", 
    canvas = intro_scene # NEED TO CHANGE AFTER YOU COPY
    )

craft = sphere(
    pos = craftloc * scalefactor,
    radius = 0.15,
    color = color.yellow,
    make_trail = True,
    canvas = intro_scene
    )
# --------------------
# initial values
# --------------------
vcraft = vector(0, 2e3, 0) # initial velocity is straight up along y
pcraft = mcraft * vcraft
```

### Part 4: Adding the while loop

In this final part, we create a while loop so that our simulation will animate. We also declare that if you hit the 'q' key, the animation will stop. Below is the framework of the while loop. Your job is to:

- First, add the radius vector: `r = craftloc - planetloc`
- Add `rmag`, `rhat`, `F`, and `Fg` just as you did in the Gravity lab
- Update the momentum: `pcraft = pcraft + Fg*dt`
- Update the display (`craft.pos`)

**That's it! Play around with changing the initial velocity to see different orbits. You may want to adjust `scalefactor` to see larger objects. If the craft tries to go through the Earth, you can add a stopping statement inside the while loop**

`if rmag < rEarth: break`

```python

# --------------------
# CALCULATION LOOP: ALL REPEATED CALCULATIONS GO INSIDE THE LOOP 
# --------------------
while t < time_limit and t > -1:
    rate(frames) # slow down motion to make animation look nicer

    # PUT YOUR CODE FOR GRAVITY HERE

    craftloc += (pcraft/mcraft) * dt

    t += dt
    
    ### Check if we want to exit ###
    k = keysdown()

    if 'q' in k:
        # can change the code to just stop the while loop
        intro_scene.delete() # delete the scene (closes the scene of the spacecraft and Earth)
        break # break out of the while loop

    
print("All done") # print a message that everything is deleted
```
