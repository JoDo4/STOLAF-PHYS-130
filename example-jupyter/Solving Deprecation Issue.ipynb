```python
# -----------------------------
# Scene setup
# -----------------------------
scene2 = canvas(title="Projectile Motion Simulation", width=800, height=600, background=color.white)
# -----------------------------
# Ground
# -----------------------------
ground = box(pos=vector(0,-0.5,0), size=vector(100,1,20), color=color.green)

# -----------------------------
# Parameters
# -----------------------------
g = vector(0, -9.8, 0)        # gravitational acceleration (m/s^2)
angle = 45                    # launch angle in degrees
speed = 20                    # initial speed (m/s)
dt = 0.01                     # time step

# Convert angle to radians
theta = radians(angle)

# Initial velocity components
v0 = vector(speed*cos(theta), speed*sin(theta), 0)

# -----------------------------
# Create projectile
# -----------------------------
ball = sphere(
    pos=vector(0,0,0),
    radius=0.5,
    color=color.red,
    make_trail=True
)

ball.velocity = v0

# -----------------------------
# Simulation loop
# -----------------------------
while ball.pos.y >= 0:
    rate(100)  # controls animation speed
    
    # Update velocity (gravity only)
    ball.velocity = ball.velocity + g*dt
    
    # Update position
    ball.pos = ball.pos + ball.velocity*dt
```

# Steps/tips for divesting from Anaconda and making Jupyter Notebook work
- Create a directory to store the virtual environment and notebooks
- Open the admin terminal and run `set-executionpolicy unrestricted`
- Close out of the admin terminal and open a normal terminal and go into the directory you created
- run `python -m venv .venv`. This creates a virtual environment that has the name ".venv"
- run `.venv\scripts\activate` this activates the virtual environment
- run `pip install jupyter ipykernel`
- run `pip install vpython`
- run `python -m ipykernel install --user --name=.venv --display-name "example-env"` This creates a kernel named "example-env"
- run `jupyter notebook` to open Jupyter Notebook in your web browser

# How to get rid of pkg_resources warning 
- Run the code `from vpython import *` and you should get an error/warning that gives you a path to an \_\_init\_\.py file.
- Open the \_\_init\_\_.py file in vscode or pycharm and you will see the top line to be `from pkg_resources import get_distribution, DistributionNotFound`
- Replace "pkg_resources" with "importlib" and replace "get_distribution,DistributionNotFound" with "metadata"
- Change line 6 to `__version__ = metadata.version(__name__)`
- Change line 7 to `except metadata.PackageNotFoundError:`
- Delete line 13 and 14, and replace with `del metadata`
- And your done!
